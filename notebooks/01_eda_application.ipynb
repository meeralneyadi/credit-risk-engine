```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_columns", 100)
plt.style.use("seaborn-v0_8")
```

```python
DATA_PATH = "../data/raw/"
FILE_NAME = "default of credit card clients.xls"

df = pd.read_excel(DATA_PATH + FILE_NAME, header=1)

print("Dataset shape:", df.shape)
df.head()
```

```python
df.columns = (
    df.columns
    .str.strip()
    .str.upper()
    .str.replace(" ", "_")
)
df.head()
```

```python
df = df.rename(columns={"DEFAULT_PAYMENT_NEXT_MONTH": "DEFAULT"})
df["DEFAULT"].value_counts()
```

```python
df.describe().T.head(10)
```

```python
default_rate = df["DEFAULT"].mean()
print(f"Default rate: {default_rate:.2%}")
```

## Final EDA Summary

- Dataset contains ~30,000 credit card clients
- Default rate is ~22%, indicating moderate class imbalance
- No missing values present
- Mostly numerical features
- Suitable for classical ML models and automated comparison
